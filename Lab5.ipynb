```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.arange(-20, 20, 1)
x2 = (123-8.2*x1)/7.9
plt.grid()
plt.plot(x1, x2, color = 'purple', linestyle='dashed');
plt.title('$8.2x_1+7.9x_2=123$', fontsize=15)
plt.xlabel('x1', fontsize=12)
plt.ylabel('x2', fontsize=12, color='black')
```

```python
import matplotlib.pyplot as plt
import numpy as np


x1 = np.arange(-42, 42, 1)
x2 = (123-8.2*x1)/7.9
x3 = (201-11.2*x1)/16.3
x4 = (169+16.3*x1)/(-8.3)
A = [[8.2, 7.9], [11.2, 16.3]]
A1 = [[11.2, 16.3], [-16.3, -8.3]]
A2 = [[-16.3, -8.3], [8.2, 7.9]]
B = [[123], [201]]
B1 = [[201], [169]]
B2 = [[169], [123]]
fig, ax = plt.subplots(figsize=(5, 5), dpi = 150)
plt.grid()
f0 = plt.plot(x1, x2, color = 'blue', linestyle='dashed');
plt.text(-14, 18, '$8.2x_1+7.9x_2=123$', fontsize = 6, rotation=-26, color='blue')
f1 = plt.plot(x1, x3, color = 'green', linestyle='dotted');
plt.text(10, -4, '$11.2x_1+16.3x_2=201$', fontsize = 6, rotation=-18, color='green')
f2 = plt.plot(x1, x4, color = 'red', linestyle='dashdot');
plt.text(-14, -16, '$-16.3x_1-8.3x_2=169$', fontsize = 6, rotation=-42, color='red')
plt.legend(['f[0]', 'f[1]', 'f[2]'])
idx = np.argwhere(np.diff(np.sign(x2 - x3))).flatten()
f0f2 = np.linalg.solve(A, B)
f1f2 = np.linalg.solve(A1, B1)
f0f1 = np.linalg.solve(A2, B2)
plt.plot(f0f2[0], f0f2[1], color = '#00a816', marker='.')
plt.annotate('The intersection point between f[0] and f[1]', fontsize = 7, xy=(f0f2[0], f0f2[1]), xytext=(f0f2[0]+30, f0f2[1]-60),
            arrowprops=dict(width = 0.5, headlength = 5, headwidth = 3, facecolor='black', shrink=0.05))
plt.plot(f1f2[0], f1f2[1], color = '#00a816', marker='.')
plt.annotate('The intersection point between f[1] and f[2]', fontsize = 7, xy=(f1f2[0], f1f2[1]), xytext=(f1f2[0]-20, f1f2[1]-78),
            arrowprops=dict(width = 0.5, headlength = 5, headwidth = 3, facecolor='black', shrink=0.05))
plt.plot(f0f1[0], f0f1[1], color = '#00a816', marker='.')

fillarayX = [f0f2[0], f1f2[0], f0f1[0]]
fillarayY = [f0f2[1], f1f2[1], f0f1[1]]
plt.fill(fillarayX, fillarayY, color='#8f8f8f', alpha=0.5)
plt.title('All lines in one figure', fontsize=15)

plt.xlabel('x1', fontsize=12)
plt.ylabel('x2', fontsize=12, color='black')
plt.savefig('my_plot.jpg')
plt.savefig('my_plot.png')
plt.savefig('my_plot.svg')
```
